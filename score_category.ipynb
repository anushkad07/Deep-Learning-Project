```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('measures_filtered.csv')
df.head()
```

```python
print(len(df))
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Final Score'], bins=20, color='skyblue', edgecolor='black')  # Adjust the number of bins as needed
plt.title('Frequency of Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Define thresholds based on quartiles
thresholds = [0.3, 0.4, 0.6, 0.8, 1]

# Assign categories based on thresholds
def assign_category(score):
    if score <= thresholds[0]:
        return "Very Poor"
    elif thresholds[0] < score <= thresholds[1]:
        return "Poor"
    elif thresholds[1] < score <= thresholds[2]:
        return "Fair"
    elif thresholds[2] < score <= thresholds[3]:
        return "Good"
    else:
        return "Very Good"

# Apply the function to create a new column with categories
df['category'] = df['Final Score'].apply(assign_category)
print(thresholds)
```

```python
# save dataframe as csv
df.to_csv('measures_categories.csv', index=False)
```

```python
import matplotlib.pyplot as plt

# Assuming your dataframe is named 'df' and it contains a column 'category' indicating the category of each score

# Group the data by 'category' and count the number of samples in each category
category_counts = df.groupby('category').size()

# Plot the histogram
plt.figure(figsize=(8, 6))
category_counts.plot(kind='bar', edgecolor='black')
plt.title('Number of Samples in Each Category')
plt.xlabel('Category')
plt.ylabel('Number of Samples')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)
plt.show()
```
